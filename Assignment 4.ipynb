1. Scrape the details of most viewed videos on YouTube from Wikipedia:

Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

You need to find following details:

    A) Rank

    B) Name

    C) Artist

    D) Upload date

    E) Views


```python
wiki = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
url=requests.get(wiki).text
soup=BeautifulSoup(url,'html.parser')
```

```python
len(soup.find_all('table'))
soup.title.string
```

```python
table_soup=soup.find_all('table')
```

```python
filters=[table for table in table_soup if table.caption is not None]

required_table=None

for table in filters:
    if str(table.caption.string).strip() == 'List of most-viewed YouTube videos':
        required_table=table
        break
```

```python
table
```

```python
headers=[header.text.strip() for header in table.find_all('th')]
headers
```

```python
rows=[]
data_rows = table.find_all('tr')

for row in data_rows:
    value=row.find_all('td')
    bfvalue=[ele.text.strip() for ele in value]
    
    if len(bfvalue)==0:
        continue
    rows.append(bfvalue)
```

```python
rows
```

```python
data = pd.DataFrame(rows)
```

```python
youtube=pd.DataFrame({})
youtube['Name']=data[0]
youtube['Author']=data[1]
youtube['Upload Date']=data[3]
youtube['Views']=data[2]
youtube
```

Url = https://www.bcci.tv/.

You need to find following details:

    A) Match title (I.e. 1st ODI)

    B) Series

    C) Place

    D) Date

    E) Time

    Note: - From bcci.tv home page you have reach to the international fixture page through code.


```python
driver = webdriver.Chrome('E:\Data trained\DataTrained_internship\Data Scraping_selenium\chromedriver')
driver.get('https://www.bcci.tv/')
time.sleep(5)
```

```python
fixture = driver.find_element_by_xpath("//a[@class = 'navigation__link navigation__link--in-drop-down']")
x = fixture.get_attribute('href')
driver.get(x)
```

```python
title = driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")
name = driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__format']")
day = driver.find_elements_by_xpath("//span[@class='fixture__date']")
mont = driver.find_elements_by_xpath("//span[@class='fixture__month']")
time = driver.find_elements_by_xpath("//span[@class='fixture__time']")
loc = driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")
```

```python
Match=[]
Series=[]
place=[]
days=[]
month=[]
times=[]
for i in title:
    Match.append(i.text)
for j in name:
    Series.append(j.text)
for k in day:
    days.append(k.text)
for l in mont:
    month.append(l.text)
for m in time:
    times.append(m.text)
for n in loc:
    place.append(n.text)
    
    
```

```python
fixture=pd.DataFrame({})
fixture['Match Title']=Match
fixture['Series']=Series
fixture['Place']=place
fixture['Day']=days
fixture['Month']=month
fixture['Time']=times
fixture
```

Scrape the details most watched tv series of all time from imdb.com.

        D) Run time

        E) Ratings

```python
driver = webdriver.Chrome('E:\Data trained\DataTrained_internship\Data Scraping_selenium\chromedriver')
driver.get('https://www.imdb.com/list/ls095964455/')
time.sleep(5)
```

```python
name=[]
nm=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
for i in nm:
    name.append(i.text)
```

```python
year=[]
yr=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
for i in yr:
    year.append(i.text)
```

```python
genre=[]
gen=driver.find_elements_by_xpath("//span[@class='genre']")
for i in gen:
    genre.append(i.text)
```

```python
run=[]
rn=driver.find_elements_by_xpath("//span[@class='runtime']")
for i in rn:
    run.append(i.text)
```

```python
votes=[]
vt=driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]")
for i in vt:
    votes.append(i.text.replace('Votes: ',''))
```
