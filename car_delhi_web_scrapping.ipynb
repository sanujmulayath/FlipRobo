```python
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
```

```python
driver=webdriver.Chrome('E:\Data trained\DataTrained_internship\Data Scraping_selenium\chromedriver.exe')
```

```python
driver.get('https://www.olx.in/delhi_g2001152/cars_c84')
time.sleep(2)
```

```python
#creating empty list
brand=[]
model=[]
variant=[]
YOM=[]
Transmission=[]
Fueltype=[] 
Kms=[]
Price=[]
Owners=[]
```

```python
Prod_url=[]

for i in range(0,700):
    url=driver.find_elements_by_xpath("//ul[@class='rl3f9 _3mXOU']/li/a")
    for i in url:
        Prod_url.append(i.get_attribute('href'))
    try:
        next_btn=driver.find_element_by_xpath("//button[@class='rui-39-wj rui-3evoE rui-1JPTg']").click()
    except NoSuchElementException:
        pass
    
for i in Prod_url:
    driver.get(i)
    time.sleep(2)
    
    try:
        brand.append(driver.find_element_by_xpath("//div[@class='_3_knn']/div/span[2]").text)
    except NoSuchElementException as e:
        brand.append('-')
        
    #getting the car model
    try:
        model.append(driver.find_element_by_xpath("//span[@data-aut-id='value_model']").text)
    except NoSuchElementException as e:
        model.append('-')
    
    #getting the car variant
    try:
        variant=[].append(driver.find_element_by_xpath("//span[@data-aut-id='value_variant']").text)
    except NoSuchElementException as e:
        variant=[].append('-')
    
    #getting the year of manufacturing
    try:
        YOM.append(driver.find_element_by_xpath("//span[@data-aut-id='value_year']").text)
    except NoSuchElementException as e:
        YOM.append('-')
    
    #getting the fuel type
    try:
        Fueltype.append(driver.find_element_by_xpath("//span[@data-aut-id='value_petrol']").text)
    except NoSuchElementException as e:
        Fueltype.append('-')
    
    #getting the transmission
    try:
        Transmission.append(driver.find_element_by_xpath("//span[@data-aut-id='value_transmission']").text)
    except NoSuchElementException as e:
        Transmission.append('-')
    
    #getting the kms run
    try:
        Kms.append(driver.find_element_by_xpath("//span[@data-aut-id='value_mileage']").text)
    except NoSuchElementException as e:
        Kms.append('-')
        
    #getting the number of owners    
    try:
        Owners.append(driver.find_element_by_xpath("//span[@data-aut-id='value_first_owner']").text)
    except NoSuchElementException as e:
        Owners.append('-')
    
    #getting the price for the car
    try:
        Price.append(driver.find_element_by_xpath("//section[@class='_2wMiF']/span[1]").text.replace('â‚¹',''))
    except NoSuchElementException as e:
        Price.append('-')
```
